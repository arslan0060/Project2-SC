```python
%pylab inline
t = loadtxt('t.txt')
x=loadtxt('x(t).txt');
y=loadtxt('y(t).txt');
plot(t, x, 'b-')
xlabel("t")
ylabel("x(t)")
title("x(t) vs t")
```

```python
plot(t, y, 'r-', label='$y(t)$')
xlabel("t")
ylabel("y(t)")
title("y(t) vs t")
```

```python
plot(x,y ,color="black")
xlabel("x(t)")
ylabel("y(t)")
title("y(t) vs x(t)")
```
